# Data Visualization

```python
#Checking column relationship with other columns,using heatmap
df_cr=df.corr().abs()
plt.figure(figsize=(12,6))
sns.heatmap(df_cr,annot=True,annot_kws={'size':8})
plt.title("Heatmap of the dataset")
plt.show()
     
```

#### 1.maximum value 0.68 observe in the indepent variable means not looks multicollinearity issue
#### 2. All values are positively correlated with each other
#### 3. Alcohol is more contribute in the target or dependent variable

```python
# Checking outliers present or not in the dataset
plt.figure(figsize=(10,6),facecolor="yellow")
plotnumber=1

for i in column_list:
    if plotnumber<=12:
        ax=plt.subplot(1,1,plotnumber)
        sns.boxplot(df[i])
        plt.xlabel (i,fontsize=12)
        plt.yticks(rotation=0, fontsize=10)
    plotnumber+=1
plt.tight_layout()
```
